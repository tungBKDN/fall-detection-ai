```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns

DATA = '../../Dataset'
ADL = DATA + '/ADL'
FALLING = DATA + '/Falling'
TARGET  = '../train-windowed-data'
```

```python
WLENGTH = 350
SEP = 300
sample = 0

def windowing(df: pd.DataFrame):
    # Split the data into windows
    global sample
    start = 0
    while start + WLENGTH < len(df):
        sample += 1
        window = df.iloc[start:start+WLENGTH]
        window.to_csv(TARGET + '/falling/' + str(sample) + '.csv', index=False)
        start += SEP

def access_files():
    # Access ADL files
    for falling in os.listdir(FALLING):
        for file in os.listdir(FALLING + '/' + falling):
            print('Reading file:', file)
            df = pd.read_csv(FALLING + '/' + falling + '/' + file, sep=';')
            windowing(df)

# df = pd.read_csv(ADL + '/ADL01/P01_01.txt', sep=';')
# df.head(20)
access_files()
```
