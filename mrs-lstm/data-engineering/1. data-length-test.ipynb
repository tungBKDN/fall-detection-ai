```python
import pandas as pd
import os

DATA_PATH = '../../Dataset'
ADL = DATA_PATH + '/ADL'
FALL = DATA_PATH + '/Falling'
```

```python
adl_lengths = []
fall_lengths = []

adl_folders = os.listdir(ADL)
for folder in adl_folders:
    files = os.listdir(ADL + '/' + folder)
    for file in files:
        df = pd.read_csv(ADL + '/' + folder + '/' + file)
        adl_lengths.append(len(df))

fall_folders = os.listdir(FALL)
for folder in fall_folders:
    files = os.listdir(FALL + '/' + folder)
    for file in files:
        df = pd.read_csv(FALL + '/' + folder + '/' + file)
        fall_lengths.append(len(df))

# Print the maximum length of the ADL and FALL datasets
print('Max Length of ADL: ', max(adl_lengths))
print('Max Length of FALL: ', max(fall_lengths))
# Print the minimum length of the ADL and FALL datasets
print('Min Length of ADL: ', min(adl_lengths))
print('Min Length of FALL: ', min(fall_lengths))
```

```python
# Distribution of the lengths of the ADL and FALL datasets
adl_lengths = pd.Series(adl_lengths)
fall_lengths = pd.Series(fall_lengths)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.distplot(adl_lengths, hist=False, label='ADL')
sns.distplot(fall_lengths, hist=False, label='FALL')
plt.xlabel('Length of the dataset')
plt.ylabel('Density')
plt.title('Distribution of the lengths of the ADL and FALL datasets')
plt.legend()
plt.show()
```

```python
# Example data
# adl_lengths = [1, 2, 3, 4, 5]
# fall_lengths = [2, 3, 4, 5, 6]

# Define boxprops
boxprops = dict(linestyle='-', linewidth=2, color='blue')

# Boxplot of the lengths of the ADL and FALL datasets
plt.figure(figsize=(10, 5))
sns.boxplot(data=[adl_lengths, fall_lengths], boxprops=boxprops)
plt.xlabel('ADL and FALL datasets')
plt.ylabel('Length of the dataset')
plt.title('Quartile values of the lengths of the ADL and FALL datasets')
plt.show()
```
